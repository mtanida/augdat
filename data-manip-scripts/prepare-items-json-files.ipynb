```python
src_path = '/Users/myazdaniUCSD/Downloads/pd-visualization-DUP/img/imgs/'
image_types = (".jpg", ".png", ".JPG", ".PNG", ".JPEG", ".jpeg",".tif", ".tiff", ".TIFF", '.TIF')
 
image_paths = []  
for root, dirs, files in os.walk(src_path):
    image_paths.extend([os.path.join(root, f) for f in files if f.endswith(image_types)])
    
print 'number of images is', len(image_paths)
```

```python
image_paths
```

```python
image_paths[0].split("/pd-visualization-DUP/")
```

```python
image_paths_rel = ["./" + image_path.split("/pd-visualization-DUP/")[1] for image_path in image_paths]
```

```python
our_items = [[image_path, "", ""] for image_path in image_paths_rel]
```

```python
our_dict = {"items": our_items, "page": 0}
```

```python
with open('our_items.json', 'w') as fp:
    json.dump(our_dict, fp)
```
